```python
import sqlite3
import pandas as pd
import os
```

```python
os.getcwd()
```

```python
connection = sqlite3.connect('E:\\GitHub\\Nillabot\\ia.db')
cursor = connection.cursor()
```

```python
gems = cursor.execute('SELECT name, attribute_label, basic_skill_id, upgrade_skill_id, icon FROM essence_gem').fetchall()
blessings = cursor.execute('SELECT name, desc, icon, quality_level FROM essence_bless').fetchall()
attribs = cursor.execute('SELECT cfg_id, name, type, quality_level, attribute_label, weight, icon FROM essence_attr').fetchall()
illusion_events = cursor.execute('SELECT name, weight FROM essence_event').fetchall()
```

```python
gem_cols = ['name', 'attribute_label', 'basic_skill_id', 'upgrade_skill_id', 'icon']
bless_cols = ['name', 'desc', 'icon', 'quality_level']
attrib_cols = ['cfg_id', 'name', 'type', 'quality_level', 'attribute_label', 'weight', 'icon']
event_cols = ['name', 'weight']
```

```python
gems_df = pd.DataFrame(gems, columns=gem_cols)
bless_df = pd.DataFrame(blessings, columns=bless_cols)
attrib_df = pd.DataFrame(attribs, columns=attrib_cols)
event_df = pd.DataFrame(illusion_events, columns=event_cols)
```
